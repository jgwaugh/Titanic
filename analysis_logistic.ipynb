```python
#script for analysis

import numpy as np
import pandas as pd
import cleaned_data as data
from sklearn import linear_model as lm


train = data.train


#logistic regresion

# now, make a make design matrix for models
# only features I believe matter are age, sex, class, and fare (maybe cabin location
# also matters?? check later)

survival = train.Survived
predictors = train[['Pclass', 'Age', 'Sex', 'Fare']]

y = survival.values
X = predictors.values



#analysis
mdl = lm.LogisticRegression()
mdl.fit(X,y)

#see how well this predicts the training data
prediction = mdl.predict(X)


summary = pd.DataFrame()
summary['true'] = y
summary['prediction'] = prediction
mispredicted = np.abs(summary.true.values - summary.prediction.values)



train['mispredicted'] = mispredicted

mistaken = train[train['mispredicted'] == 1]



```

```python
mdl.score(X,y)
```
